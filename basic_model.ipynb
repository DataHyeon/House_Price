# Data Load

```python
cwd = os.getcwd()
```

## MinMax Data

```python
mm_train_path = glob(f'{cwd}\\**\\mm_train.csv', recursive=True)[0]
mm_test_path = glob(f'{cwd}\\**\\mm_test.csv', recursive=True)[0]

print(mm_test_path, mm_test_path, sep='\n\n')
```

```python
mm_train = pd.read_csv(mm_train_path)
mm_test = pd.read_csv(mm_test_path)
```

## Standard Data

```python
st_train_path = glob(f'{cwd}\\**\\st_train.csv', recursive=True)[0]
st_test_path = glob(f'{cwd}\\**\\st_test.csv', recursive=True)[0]

print(st_test_path, st_test_path, sep='\n\n')
```

```python
st_train = pd.read_csv(st_train_path)
st_test = pd.read_csv(st_test_path)
```

# Pycaret - MM

```python
c_mm = mm_train.copy()
c_mm = c_mm.iloc[:, 1:]
```

```python
mm_caret = setup(data=c_mm, target='SalePrice', fold=5, train_size=0.7, session_id=1234)
```

```python
mm_comp = mm_caret.compare_models(n_select=5)
mm_result = mm_caret.pull()
mm_result
```

## Tune -MM

```python
cat_tune = mm_caret.tune_model(mm_comp[0])
```

```python
tune_result = mm_caret.pull()
tune_result
```

```python
pred = mm_caret.predict_model(cat_tune, data=mm_test.iloc[:,1:])
```

```python
submission = pd.read_csv(r'C:\Users\reot1\OneDrive\바탕 화면\House Prices\dataset\sample_submission.csv')
submission['SalePrice'] = pred.iloc[:,-1]
submission.to_csv('./min_max.csv', index=False)
```

# Pycaret -St

```python
c_st = st_train.copy()
c_st = c_st.iloc[:,1:]
```

```python
st_caret = setup(data=c_st, target='SalePrice', fold=5, train_size=0.7, session_id=1234)
```

```python
st_comp = st_caret.compare_models(n_select=5)
st_result = st_caret.pull()
st_result
```

## Tune-St

```python
st_cat_tune = st_caret.tune_model(st_comp[0])
st_tune_result = st_caret.pull()
st_tune_result
```

```python
st_pred = st_caret.predict_model(st_cat_tune, data=st_test.iloc[:,1:])
```

```python
submission = pd.read_csv(r'C:\Users\reot1\OneDrive\바탕 화면\House Prices\dataset\sample_submission.csv')
submission['SalePrice'] = st_pred.iloc[:,-1]
submission.to_csv('./standard.csv', index=False)
```
